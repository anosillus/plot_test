```python
from tabula import wrapper
pre_checkup_number_with_prefecture = wrapper.read_pdf(URL_OF_PDF_FILE_WITH_CHECKUP_NUMBER_WITH_PREFECTURE, pages = 1, encoding = "utf-8_sig", spreadsheet=True)

not_index_row = pre_checkup_number_with_prefecture.iloc[:,0] != '都道府県名'
not_sum_row = pre_checkup_number_with_prefecture.iloc[:,0] != '合計'
not_anotation_row = ~pre_checkup_number_with_prefecture.iloc[:,0].str.startswith('※')

checkup_number_with_prefecture = pre_checkup_number_with_prefecture[not_sum_row & not_index_row & not_anotation_row].drop("備考", axis=1)
if len(checkup_number_with_prefecture) != COUNT_OF_PREFECTURES:
    print("ERROR")
else:
    print("(Maybe) Good DATA!")
```
